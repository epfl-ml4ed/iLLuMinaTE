```python
import pandas as pd

course="geomatique_003"
```

```python
index=pd.read_csv("../easy-fail/eq_week-chen_cui-" + str(course) + "/feature_labels.csv",)

# Make the index a column called 'feature_index'
index.reset_index(inplace=True)
index.rename(columns={'index': 'feature_index'}, inplace=True)

# Rename 'Unnamed: 0' to 'original_index'
index.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)

# Remove the other columns except 'feature_index' and 'original_index'
index = index[['feature_index', 'original_index']]
index
```

```python
# Load data and filter the easy fail students
data = pd.read_csv(f"data/pattern_labels-{course}.csv")
data = data[data["Unnamed: 0"].isin(index["original_index"])]
data
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np



# Assuming the user_id is not a feature for clustering
features = data.drop(['user_id'], axis=1)

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Number of clusters
n_clusters = 5  # This is set to 20 as we want 20 representative samples

# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(features_scaled)

# Add cluster labels to the dataframe
data['cluster'] = clusters

# Select a representative sample from each cluster
representative_samples = data.groupby('cluster').apply(lambda x: x.sample(2)).reset_index(drop=True)

# Drop the cluster column and return only the representative samples
representative_samples = representative_samples.drop('cluster', axis=1)
```
