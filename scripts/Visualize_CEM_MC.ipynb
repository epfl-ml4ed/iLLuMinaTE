```python
from PIL import Image
import os

def resize_images_in_folder(folder_path, target_height):
    # Create a directory for resized images
    resized_folder = os.path.join(folder_path, 'resized')
    os.makedirs(resized_folder, exist_ok=True)

    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)

        # Check if it's a file and an image
        if os.path.isfile(file_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            with Image.open(file_path) as img:
                # Calculate the new width to maintain the aspect ratio
                aspect_ratio = img.width / img.height
                new_width = int(target_height * aspect_ratio)

                # Resize the image
                resized_img = img.resize((new_width, target_height), Image.ANTIALIAS)

                # Save the resized image in the 'resized' folder
                resized_img.save(os.path.join(resized_folder, filename))

                print(f"Resized {filename} to {new_width}x{target_height}")

folder_path = '../'  # Replace with the path to your folder
target_height = 500  # Replace with the desired height

resize_images_in_folder('gpt_visuals/gpt_visuals_new/', target_height)
```
